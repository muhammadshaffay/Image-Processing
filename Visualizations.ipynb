#### Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Load the dataset of medical images

```python
png_images = ['./png/Image1.png', './png/Image2.png', './png/Image3.png', './png/Image4.png']
```

```python
def ImageReader(images):
    
    # Reader
    reader = []
    for i in images:
        img = cv2.imread(i)
        reader.append(img)
        
    return reader
```

```python
def displayer(images):
    
    # Displayer
    for img in images:
        plt.imshow(img)
        plt.show()
```

```python
images = ImageReader(png_images)
displayer(images)
```

## Functions to perform Visualizations

#### Image histogram

```python
def histogram_image(image):

    # Plot the histogram
    plt.hist(image.ravel(), 256, [0, 256])
    plt.xlabel('Intensity Value')
    plt.ylabel('Pixel Count')
    plt.title('Image Histogram')
    plt.savefig('Histogram.png')
    plt.show()
```

```python
histogram_image(images[1])
```

#### Image gradient magnitude

```python
def gradient_magnitude_image(image):
    
    # Calculate the gradient magnitude using the Sobel operator
    grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    
    gradient_magnitude = np.sqrt((grad_x**2) + (grad_y**2))
    
    # display
    plt.imshow(gradient_magnitude)
    plt.show()
    
    return gradient_magnitude
```

```python
gradient_magnitude = gradient_magnitude_image(images[1])
cv2.imwrite('Gradient Magnitude.png', gradient_magnitude)
```

#### Image thresholding

```python
def threshold_image(image, threshold):
    
    # binary thresholding 
    ret, thresh = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
    
    # display
    plt.imshow(thresh)
    plt.show()
    
    return thresh
```

```python
threshold = threshold_image(images[1], 128)
cv2.imwrite('Thresold.png', threshold)
```
